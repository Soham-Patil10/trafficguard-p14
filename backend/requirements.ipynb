## Cell 2 — Verify all packages imported correctly

```python
import importlib

checks = {
    'fastapi'          : 'fastapi',
    'uvicorn'          : 'uvicorn',
    'python-multipart' : 'multipart',
    'torch'            : 'torch',
    'torchvision'      : 'torchvision',
    'Pillow'           : 'PIL',
    'numpy'            : 'numpy',
    'pandas'           : 'pandas',
    'tqdm'             : 'tqdm',
    'scikit-learn'     : 'sklearn',
}

all_ok = True
for pkg_name, import_name in checks.items():
    try:
        mod = importlib.import_module(import_name)
        version = getattr(mod, '__version__', 'unknown')
        print(f'  ✓  {pkg_name:<20} {version}')
    except ImportError:
        print(f'  ✗  {pkg_name:<20} NOT FOUND')
        all_ok = False

print()
if all_ok:
    print('✅ All packages installed successfully — open main.ipynb to start the server.')
else:
    print('❌ Some packages are missing — re-run Cell 1.')
```
